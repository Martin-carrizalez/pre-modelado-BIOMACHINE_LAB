```python
# CELDA 10: Funci√≥n para analizar outliers con m√©todo IQR
def analyze_outliers(df, column_name):
    """
    Detecta outliers usando el m√©todo IQR (Interquartile Range)
    
    IQR = Q3 - Q1
    Outlier moderado: fuera de [Q1 - 1.5*IQR, Q3 + 1.5*IQR]
    Outlier extremo: fuera de [Q1 - 3*IQR, Q3 + 3*IQR]
    """
    # Filtrar valores no nulos
    data = df[column_name].dropna()
    
    if len(data) == 0:
        return None
    
    # Calcular cuartiles
    q1 = data.quantile(0.25)
    q2 = data.quantile(0.50)  # mediana
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    
    # L√≠mites para outliers moderados (1.5 * IQR)
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # L√≠mites para outliers extremos (3 * IQR)
    lower_extreme = q1 - 3 * iqr
    upper_extreme = q3 + 3 * iqr
    
    # Identificar outliers
    outliers_mask = (data < lower_bound) | (data > upper_bound)
    extreme_mask = (data < lower_extreme) | (data > upper_extreme)
    
    n_outliers = outliers_mask.sum()
    n_extreme = extreme_mask.sum()
    
    # Estad√≠sticas
    mean = data.mean()
    std = data.std()
    
    return {
        'variable': column_name,
        'n': len(data),
        'mean': mean,
        'std': std,
        'min': data.min(),
        'q1': q1,
        'median': q2,
        'q3': q3,
        'max': data.max(),
        'iqr': iqr,
        'lower_bound': lower_bound,
        'upper_bound': upper_bound,
        'lower_extreme': lower_extreme,
        'upper_extreme': upper_extreme,
        'n_outliers': n_outliers,
        'outliers_pct': (n_outliers / len(data)) * 100,
        'n_extreme': n_extreme,
        'extreme_pct': (n_extreme / len(data)) * 100
    }

print("‚úÖ Funci√≥n de an√°lisis creada correctamente")
```

```python
# CELDA 11 - VERSI√ìN MEJORADA: Separar tipos de variables
print("="*80)
print("üîç AN√ÅLISIS DE OUTLIERS - CLASIFICACI√ìN INTELIGENTE DE VARIABLES")
print("="*80)

# TIPO 1: Variables continuas REALES (analizar outliers con IQR)
continuous_vars = [
    ' Age (yrs)',
    'Weight (Kg)',
    'Height(Cm) ',
    'BMI',
    'Pulse rate(bpm) ',
    'RR (breaths/min)',
    'Hb(g/dl)',
    'Hip(inch)',
    'Waist(inch)',
    'Waist:Hip Ratio',
    'FSH(mIU/mL)',
    'LH(mIU/mL)',
    'FSH/LH',
    'TSH (mIU/L)',
    'AMH(ng/mL)',
    'PRL(ng/mL)',
    'Vit D3 (ng/mL)',
    'PRG(ng/mL)',
    'RBS(mg/dl)',
    'BP _Systolic (mmHg)',
    'BP _Diastolic (mmHg)',
    '  I   beta-HCG(mIU/mL)',
    'II    beta-HCG(mIU/mL)',
    'Avg. F size (L) (mm)',
    'Avg. F size (R) (mm)',
    'Endometrium (mm)'
]

# TIPO 2: Variables de CONTEO (revisar pero no aplicar IQR estricto)
count_vars = [
    'No. of abortions',
    'Follicle No. (L)',
    'Follicle No. (R)',
    'Cycle length(days)',
    'Marraige Status (Yrs)'
]

# TIPO 3: Variables categ√≥ricas/binarias (NO analizar)
categorical_vars = [
    'PCOS (Y/N)',
    'Blood Group',
    'Cycle(R/I)',
    'Pregnant(Y/N)',
    'Weight gain(Y/N)',
    'hair growth(Y/N)',
    'Skin darkening (Y/N)',
    'Hair loss(Y/N)',
    'Pimples(Y/N)',
    'Fast food (Y/N)',
    'Reg.Exercise(Y/N)'
]

print(f"\nüìä CLASIFICACI√ìN DE VARIABLES:")
print(f"üìã Variables CATEG√ìRICAS excluidas: {len(categorical_vars)}")
print("\nüö´ Variables categ√≥ricas excluidas del an√°lisis de outliers:")
for var in categorical_vars:
    print(f"   ‚Ä¢ {var}")
print(f"   ‚úÖ Variables CONTINUAS: {len(continuous_vars)} (analizar outliers con IQR)")
print(f"   ‚ö†Ô∏è  Variables de CONTEO: {len(count_vars)} (revisar manualmente)")
print(f"   ‚ùå Variables CATEG√ìRICAS: {len(categorical_vars)} (no analizar)")

print("\n‚ö†Ô∏è  VARIABLES DE CONTEO (no usar IQR estricto):")
for var in count_vars:
    print(f"   ‚Ä¢ {var}")

# Analizar SOLO variables continuas verdaderas
all_results = []

print("\nüîç Analizando variables CONTINUAS...\n")

for col in continuous_vars:
    if col in df.columns:
        stats = analyze_outliers(df, col)
        if stats:
            all_results.append(stats)

# Crear DataFrame con resultados
results_df = pd.DataFrame(all_results)
results_df_sorted = results_df.sort_values('n_extreme', ascending=False)

# Mostrar resumen
print("\nüìä RESUMEN DE OUTLIERS (solo variables continuas):\n")
print(f"{'Variable':<35} {'Outliers':<12} {'Extremos':<12} {'% Outliers':<12}")
print("-"*75)

for _, row in results_df_sorted.iterrows():
    if row['n_outliers'] > 0:
        print(f"{row['variable']:<35} {row['n_outliers']:<12.0f} {row['n_extreme']:<12.0f} {row['outliers_pct']:<12.1f}%")

print("-"*75)
print(f"{'TOTAL:':<35} {results_df['n_outliers'].sum():<12.0f} {results_df['n_extreme'].sum():<12.0f}")

# AN√ÅLISIS SEPARADO: Variables de conteo (solo mostrar distribuci√≥n)
print("\n" + "="*80)
print("‚ö†Ô∏è  AN√ÅLISIS ESPECIAL: VARIABLES DE CONTEO")
print("="*80)

for var in count_vars:
    if var in df.columns:
        print(f"\nüìä {var}:")
        value_counts = df[var].value_counts().sort_index()
        print(f"   Distribuci√≥n de valores:")
        for val, count in value_counts.head(10).items():
            pct = (count / len(df)) * 100
            print(f"      {val}: {count} pacientes ({pct:.1f}%)")
        
        if len(value_counts) > 10:
            print(f"      ... y {len(value_counts) - 10} valores m√°s")
        
        print(f"   M√°ximo: {df[var].max()}")
        print(f"   Media: {df[var].mean():.2f}")

print(f"\n‚úÖ An√°lisis completado:")
print(f"   ‚Ä¢ {len(continuous_vars)} variables continuas analizadas con IQR")
print(f"   ‚Ä¢ {len(count_vars)} variables de conteo revisadas por distribuci√≥n")
```

```python
# CELDA 12: An√°lisis detallado de variables hormonales (las m√°s cr√≠ticas)
print("="*80)
print("üîç AN√ÅLISIS DETALLADO: VARIABLES HORMONALES CON OUTLIERS EXTREMOS")
print("="*80)

# Variables hormonales cr√≠ticas
hormonal_vars = [
    'FSH(mIU/mL)',
    'LH(mIU/mL)',
    'FSH/LH',
    'TSH (mIU/L)',
    'AMH(ng/mL)',
    'PRL(ng/mL)',
    'RBS(mg/dl)',
    '  I   beta-HCG(mIU/mL)',
    'II    beta-HCG(mIU/mL)'
]

print("\nüìä Analizando cada variable hormonal...\n")

for var in hormonal_vars:
    if var in df.columns:
        stats = analyze_outliers(df, var)
        
        if stats and stats['n_extreme'] > 0:
            print(f"\n{'='*70}")
            print(f"üî¥ {var}")
            print(f"{'='*70}")
            
            print(f"\n   üìà ESTAD√çSTICAS:")
            print(f"      N: {stats['n']} | Media: {stats['mean']:.2f} | Std: {stats['std']:.2f}")
            print(f"      Min: {stats['min']:.2f} | Mediana: {stats['median']:.2f} | Max: {stats['max']:.2f}")
            
            print(f"\n   üéØ L√çMITES (M√©todo IQR):")
            print(f"      Rango normal: [{stats['lower_bound']:.2f}, {stats['upper_bound']:.2f}]")
            print(f"      L√≠mites extremos: [{stats['lower_extreme']:.2f}, {stats['upper_extreme']:.2f}]")
            
            print(f"\n   ‚ö†Ô∏è  OUTLIERS DETECTADOS:")
            print(f"      Total outliers: {stats['n_outliers']} ({stats['outliers_pct']:.1f}%)")
            print(f"      üî¥ Outliers EXTREMOS: {stats['n_extreme']} ({stats['extreme_pct']:.1f}%)")
            
            # Encontrar y mostrar outliers extremos espec√≠ficos
            data = df[var].dropna()
            q1 = data.quantile(0.25)
            q3 = data.quantile(0.75)
            iqr = q3 - q1
            lower_extreme = q1 - 3 * iqr
            upper_extreme = q3 + 3 * iqr
            
            extreme_mask = (df[var] < lower_extreme) | (df[var] > upper_extreme)
            extreme_data = df[extreme_mask][['PCOS (Y/N)', var]].copy()
            
            if len(extreme_data) > 0:
                extreme_data['abs_value'] = extreme_data[var].abs()
                extreme_data_sorted = extreme_data.sort_values('abs_value', ascending=False)
                
                print(f"\n   üö® TOP 5 VALORES M√ÅS EXTREMOS:")
                for i, (idx, row) in enumerate(extreme_data_sorted.head(5).iterrows(), 1):
                    pcos_status = '‚úì SOP' if row['PCOS (Y/N)'] == 1 else '‚úó No-SOP'
                    excel_row = idx + 2
                    print(f"      {i}. Fila {excel_row}: {row[var]:>10.2f} [{pcos_status}]")

print("\n" + "="*80)
print("‚úÖ AN√ÅLISIS HORMONAL COMPLETADO")
print("="*80)
```

```python
# CELDA 13: Identificar y listar TODAS las filas con outliers extremos
print("\n" + "="*80)
print("üìù REPORTE: FILAS CON OUTLIERS EXTREMOS")
print("="*80)

# Lista para almacenar filas problem√°ticas
problematic_rows = []

# Analizar cada variable continua
for var in continuous_vars:
    if var in df.columns:
        data = df[var].dropna()
        
        if len(data) > 0:
            q1 = data.quantile(0.25)
            q3 = data.quantile(0.75)
            iqr = q3 - q1
            lower_extreme = q1 - 3 * iqr
            upper_extreme = q3 + 3 * iqr
            
            # Encontrar outliers extremos
            extreme_mask = (df[var] < lower_extreme) | (df[var] > upper_extreme)
            extreme_indices = df[extreme_mask].index.tolist()
            
            for idx in extreme_indices:
                problematic_rows.append({
                    'pandas_index': idx,
                    'excel_row': idx + 2,
                    'variable': var,
                    'value': df.loc[idx, var],
                    'pcos': 'SOP' if df.loc[idx, 'PCOS (Y/N)'] == 1 else 'No-SOP',
                    'lower_limit': lower_extreme,
                    'upper_limit': upper_extreme
                })

# Crear DataFrame
problematic_df = pd.DataFrame(problematic_rows)

print(f"\nüìä Total de outliers extremos detectados: {len(problematic_df)}")

# Agrupar por fila para ver qu√© filas tienen m√∫ltiples outliers
if len(problematic_df) > 0:
    rows_with_multiple = problematic_df.groupby('excel_row').size().sort_values(ascending=False)
    
    print(f"\nüîç FILAS CON M√öLTIPLES OUTLIERS EXTREMOS:")
    print(f"\n{'Fila Excel':<12} {'# Outliers':<12} {'PCOS':<10}")
    print("-"*40)
    
    for excel_row in rows_with_multiple.head(10).index:
        row_data = problematic_df[problematic_df['excel_row'] == excel_row]
        n_outliers = len(row_data)
        pcos_status = row_data.iloc[0]['pcos']
        
        print(f"{excel_row:<12} {n_outliers:<12} {pcos_status:<10}")
        
        # Mostrar qu√© variables son outliers en esa fila
        for _, item in row_data.iterrows():
            print(f"   ‚Ä¢ {item['variable']}: {item['value']:.2f}")
    
    # Filas m√°s problem√°ticas (top 3)
    print(f"\nüö® TOP 3 FILAS M√ÅS PROBLEM√ÅTICAS:")
    top_3_rows = rows_with_multiple.head(3)
    
    for i, (excel_row, n_outliers) in enumerate(top_3_rows.items(), 1):
        row_idx = excel_row - 2
        print(f"\n   {i}. FILA {excel_row}:")
        print(f"      ‚Ä¢ Outliers extremos: {n_outliers}")
        print(f"      ‚Ä¢ PCOS: {df.loc[row_idx, 'PCOS (Y/N)']}")
        
        row_outliers = problematic_df[problematic_df['excel_row'] == excel_row]
        for _, item in row_outliers.iterrows():
            print(f"      ‚Ä¢ {item['variable']}: {item['value']:.2f}")

# Guardar reporte
problematic_df.to_csv('outliers_extremos_detalle.csv', index=False)
print(f"\nüíæ Reporte guardado: outliers_extremos_detalle.csv")
```

```python
# CELDA 16: Implementar limpieza de datos seg√∫n resumen ejecutivo
print("\n" + "="*80)
# 1. Cargar ORIGINAL
df = pd.read_excel('../data/PCOS_data_1.xlsx')
print(f"Original: {len(df)} filas")

# 2. Identificar outliers cr√≠ticos
outliers_to_remove = []

# FSH > 1000
outliers_to_remove.extend(df[df['FSH(mIU/mL)'] > 1000].index.tolist())

# LH > 1000  
outliers_to_remove.extend(df[df['LH(mIU/mL)'] > 1000].index.tolist())

# TSH > 50
outliers_to_remove.extend(df[df['TSH (mIU/L)'] > 50].index.tolist())

# Eliminar duplicados
outliers_to_remove = list(set(outliers_to_remove))

print(f"\nüóëÔ∏è  Eliminando {len(outliers_to_remove)} filas:")
for idx in outliers_to_remove:
    fsh = df.loc[idx, 'FSH(mIU/mL)']
    lh = df.loc[idx, 'LH(mIU/mL)']
    tsh = df.loc[idx, 'TSH (mIU/L)']
    print(f"   Fila {idx+2}: FSH={fsh}, LH={lh}, TSH={tsh}")

# 3. Crear versi√≥n limpia
df_clean = df.drop(outliers_to_remove)
print(f"\n‚úÖ Limpio: {len(df_clean)} filas")

# 4. Guardar
df_clean.to_csv('PCOS_data_clean.csv', index=False)
print("üíæ Guardado: PCOS_data_clean.csv")
```
